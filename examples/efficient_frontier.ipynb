```python
import yapo as y

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
matplotlib.rcParams.update({'font.size': 13, 'figure.figsize': (18, 7), 'lines.markersize': 9})
```

```python
asset_names = ['us/SPY', 'us/VB', 'us/BIV', 'us/EDV', 'us/VNQ', 'us/VGK', 'us/GLD']
```

```python
start_period = '2008-2'
end_period = '2017-8'
currency = 'usd'

efficiency_frontier = y.optimal_portfolios(asset_names=asset_names,
                                           start_period=start_period, end_period=end_period, currency=currency,
                                           samples_count=20)
```

```python
[np.round(p.weights, 2) for p in efficiency_frontier.points]
```

```python
risks_yearly = []
returns_yearly = []
points = sorted(efficiency_frontier.points, key=lambda x: x.return_)[1:-1]

for pnt in points:
    risks_yearly.append(pnt.risk_yearly)
    returns_yearly.append(pnt.return_yearly)
```

```python
plt.title('Efficient frontier\nAssets: {}\nPeriod range: {} .. {}\nCurrency: {}'
              .format(', '.join(asset_names), start_period, end_period, currency))
plt.xlabel('Risks (yearly)')
plt.ylabel('Returns (yearly)')
_ = plt.plot(risks_yearly, returns_yearly, 'y-o')
```
