```python
import yapo as y

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.figsize"] = (15., 9.)
```

```python
asset_names = ['ny/SPY', 'ny/VB', 'ny/BIV', 'ny/EDV', 'ny/VNQ', 'ny/VGK', 'ny/GLD']
opt_portfolios = y.optimal_portfolios(asset_names=asset_names,
                                      start_period='2008-2', end_period='2017-8', currency='usd',
                                      samples_count=30)
```

```python
risks = [p.risk().value for p in opt_portfolios]
returns = [np.mean(p.rate_of_return().values) for p in opt_portfolios]

plt.ylabel('mean')
plt.xlabel('std')
_ = plt.plot(risks, returns, 'y-o')
```
